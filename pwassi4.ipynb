```python
# Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
"""ass:
A class is a blueprint or template for creating objects. It defines the properties (attributes) and behaviors (methods) that the objects created from the class can have. It does not hold any data by itself but provides a structure.

Key Points:
Defines attributes and methods.
Acts as a template for creating objects.
Does not consume memory until an object is created.
Object:
An object is an instance of a class. It represents a real-world entity and has its own unique data stored in the attributes defined by its class. Objects can interact with one another and invoke methods defined in their class.

Key Points:
Created from a class.
Contains actual data.
Can access the methods and attributes defined by its class.
Example:
Letâ€™s consider a simple example of a Car."""


class Car:
    def __init__(self, brand, model, color):
        self.brand = brand
        self.model = model
        self.color = color

    def display_details(self):
        print(f"Car Details: {self.color} {self.brand} {self.model}")

    def drive(self):
        print(f"The {self.color} {self.brand} {self.model} is now driving.")

car1 = Car("Toyota", "Corolla", "Red")
car2 = Car("Honda", "Civic", "Blue")

car1.display_details()  
car1.drive()            

car2.display_details()  
car2.drive()            
```

```python
"""
1. Encapsulation
Encapsulation is the concept of bundling data (attributes) and methods (functions) that operate on the data into a single unit, typically a class.
It also involves restricting direct access to some of the object's components, protecting the integrity of the data.
Example: Use of private variables and getter/setter methods in a class.
2. Inheritance
Inheritance allows one class (child/subclass) to inherit attributes and methods from another class (parent/superclass).
It promotes code reuse and establishes a relationship between the parent and child classes.
Example: A Dog class inheriting from an Animal class.
3. Polymorphism
Polymorphism means "many forms" and allows objects to be treated as instances of their parent class, while still behaving differently based on the specific object type.
This can be achieved through method overriding and overloading.
Example: A Shape class with a method area(), and subclasses like Circle and Rectangle overriding the method.
4. Abstraction
Abstraction focuses on exposing only the essential details of an object while hiding the implementation details.
This is often implemented using abstract classes or interfaces.
Example: A Vehicle abstract class defining a method move(), while specific classes like Car and Bike implement the method differently."""
```

```python
# Q3
"""The __init__() function in Python is a special method (also called a constructor) used to initialize an object when it is created. It allows setting up the initial state of the object by defining and initializing attributes.

Key Points:
Automatically called when a new object is created.
Used to initialize attributes of the class.
Helps in passing arguments during object creation.
"""
class Person:
    def __init__(self, name, age):
        self.name = name 
        self.age = age    

    def display_info(self):
        print(f"My name is {self.name} and I am {self.age} years old.")

person1 = Person("Alice", 30)
person1.display_info()  
```

```python
#Q-4
"""The self keyword in Python is used to represent the instance of the class. It allows accessing the attributes and methods of the class in object-oriented programming.

Key Points:
It refers to the current object of the class.
self must be explicitly passed as the first parameter in instance methods.
It is not a keyword; you can use any name, but self is the convention.
Example:
python
Copy code"""
class Animal:
    def __init__(self, species):
        self.species = species  

    def speak(self):
        print(f"The {self.species} makes a sound.")

# Create an object
dog = Animal("Dog")
dog.speak() 

```

```python
#Q-5

Inheritance is a fundamental concept in object-oriented programming that allows a class (child/subclass) to inherit properties and methods from another class (parent/superclass). This promotes code reuse and hierarchical classification.

Types of Inheritance and Examples:
Single Inheritance

A child class inherits from one parent class.
python
Copy code
class Parent:
    def display(self):
        print("This is the parent class.")

class Child(Parent):
    def show(self):
        print("This is the child class.")

obj = Child()
obj.display()  # Output: This is the parent class.
obj.show()     # Output: This is the child class.
Multiple Inheritance

A child class inherits from more than one parent class.
python
Copy code
class Father:
    def skill(self):
        print("Father: Good at painting.")

class Mother:
    def talent(self):
        print("Mother: Good at singing.")

class Child(Father, Mother):
    pass

obj = Child()
obj.skill()   # Output: Father: Good at painting.
obj.talent()  # Output: Mother: Good at singing.
Multilevel Inheritance

A class inherits from another class, and another class inherits from it.
python
Copy code
class Grandparent:
    def greet(self):
        print("Hello from Grandparent.")

class Parent(Grandparent):
    pass

class Child(Parent):
    pass

obj = Child()
obj.greet()  # Output: Hello from Grandparent.
Hierarchical Inheritance

Multiple child classes inherit from a single parent class.
python
Copy code
class Animal:
    def speak(self):
        print("Animal speaks.")

class Dog(Animal):
    def bark(self):
        print("Dog barks.")

class Cat(Animal):
    def meow(self):
        print("Cat meows.")

dog = Dog()
dog.speak()  # Output: Animal speaks.
dog.bark()   # Output: Dog barks.

cat = Cat()
cat.speak()  # Output: Animal speaks.
cat.meow()   # Output: Cat meows.
Hybrid Inheritance

A combination of two or more types of inheritance.
python
Copy code
class Base:
    def greet(self):
        print("Hello from Base class.")

class A(Base):
    pass

class B(Base):
    pass

class C(A, B):
    pass

obj = C()
obj.greet()  # Output: Hello from Base class.
```
